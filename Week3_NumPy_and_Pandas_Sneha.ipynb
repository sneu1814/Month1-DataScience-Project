Introduction

In this week, I explored NumPy and Pandas, which are important libraries used in data science. NumPy helps in working with numerical data efficiently, while Pandas is useful for handling structured data such as tables. These tools make it easier to clean, analyze, and process data.

Importing Libraries

First, the required libraries are imported. NumPy is used for numerical operations, and Pandas is used for data manipulation.

```python
import numpy as np
import pandas as pd
```

NumPy Array

A NumPy array is used to store numerical values. It allows faster and easier mathematical operations compared to normal Python lists.

```python
numbers = np.array([10, 20, 30, 40, 50])

print("NumPy Array:", numbers)

average_value = np.mean(numbers)

print("Average value:", average_value)
```

Pandas DataFrame

A DataFrame is a table-like structure used to store data in rows and columns. It is commonly used for analyzing datasets.

```python
data = {
    "Name": ["Sidhi", "Sneha", "Shraddha", "Simran"],
    "Marks": [85, 90, 78, 92]
}

df = pd.DataFrame(data)

print(df)
```

Data Analysis

Basic analysis can be performed using Pandas, such as calculating average, maximum, and minimum values.

```python
average_marks = df["Marks"].mean()

print("Average Marks:", average_marks)

max_marks = df["Marks"].max()

print("Maximum Marks:", max_marks)
```

Data Cleaning

Data cleaning is an important step where incorrect or missing values are handled. This improves data quality.

```python
data_with_missing = {
        "Name": ["Sidhi", "Sneha", "Shradhha", "Simran"],
    "Marks": [85, None, 78, 92]
}

df2 = pd.DataFrame(data_with_missing)

print("Before cleaning:")
print(df2)

cleaned_df = df2.dropna()

print("After cleaning:")
print(cleaned_df)
```

Mini Project: Student Marks Analysis

In this project, I created a dataset of student marks and performed analysis such as calculating average marks and identifying overall performance. This demonstrates how Pandas can be used for basic data analysis.

```python
student_data = {
    "Student": ["Shradhha", "Sidhi", "Sneha", "Simran"],
    "Marks": [88, 91, 69, 95]
}

student_df = pd.DataFrame(student_data)

print("Student Data:")
print(student_df)

avg = student_df["Marks"].mean()

print("Average Marks:", avg)

highest = student_df["Marks"].max()

print("Highest Marks:", highest)
```
