Introduction

In this week, I learned about different Python data structures such as lists, tuples, dictionaries, and sets. These structures help in storing and managing data in an organized way. I also learned about functions, which help in making the code reusable and more efficient.

These concepts are important in data science because data is usually stored and processed using these structures.

Lists

A list is a collection of multiple items stored in a single variable. Lists are ordered and can be modified.

```python
numbers = [10, 20, 30, 40, 50]

print("List values:", numbers)

print("First element:", numbers[0])
```

Tuples

A tuple is similar to a list, but it cannot be changed after creation. It is useful for storing fixed data.

```python
student_info = ("Sneha", 22, "Data Science")

print("Student Info:", student_info)
```

Dictionaries

A dictionary stores data in key-value pairs. It is useful when we want to associate values with specific keys.

```python
student = {
    "name": "Sneha",
    "age": 22,
    "course": "Data Science"
}

print("Student Name:", student["name"])
print("Course:", student["course"])
```

Sets

A set is a collection of unique elements. It automatically removes duplicate values.

```python
values = [10, 20, 20, 30, 30, 40]

unique_values = set(values)

print("Unique values:", unique_values)
```

Functions

A function is a block of code that performs a specific task. Functions help in organizing code and making it reusable.

```python
def calculate_square(number):
    result = number * number
    return result

output = calculate_square(6)

print("Square is:", output)
```

Mini Project: Basic Data Cleaning

In this project, I worked on cleaning a list of data. The program removes duplicate values and filters the data based on a condition. This is useful in real-world data processing where raw data often contains duplicate or unwanted values.

```python
data_values = [15, 25, 25, 40, 55, 55, 70, 80]

print("Original data:", data_values)

unique_data = list(set(data_values))

print("After removing duplicates:", unique_data)

filtered_data = []

for value in unique_data:
    if value > 50:
        filtered_data.append(value)

print("Filtered data (greater than 50):", filtered_data)
```
