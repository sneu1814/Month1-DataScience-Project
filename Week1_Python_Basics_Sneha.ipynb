Mini Project: Average Temperature Calculation

This program calculates the average temperature from a list of values. This simulates basic data processing which is useful in real-world data analysis.

```python
temperatures = [34, 31, 36, 29, 33, 34, 30]

total = sum(temperatures)
count = len(temperatures)

average = total / count

print("Average Temperature is:", average)
```
