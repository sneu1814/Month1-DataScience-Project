Importing Required Libraries

Matplotlib and Seaborn are used for creating visualizations. These libraries provide different types of charts for analyzing data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Line Plot

A line plot is used to show trends over time or sequence. It connects data points with a line.

```python
days = [1, 2, 3, 4, 5]
temperature = [30, 32, 31, 33, 34]

plt.plot(days, temperature)

plt.title("Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.show()
```

Bar Chart

A bar chart is used to compare values between different categories.

```python
students = ["Sidhi", "Sneha", "Shradhha", "Simran"]
marks = [85, 90, 78, 92]

plt.bar(students, marks)

plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

Histogram

A histogram shows the distribution of numerical data. It helps in understanding how values are spread.

```python
marks_data = [60, 70, 75, 80, 85, 90, 95, 70, 85, 90]

plt.hist(marks_data)

plt.title("Marks Distribution")

plt.show()
```

Heatmap

A heatmap is used to visualize relationships between variables. It uses colors to represent values.

```python
data = np.random.rand(5, 5)

sns.heatmap(data)

plt.title("Sample Heatmap")

plt.show()
```

Mini Project: Student Performance Visualization

In this project, I created visualizations to analyze student marks. Charts help in comparing performance and identifying trends.

```python
student_data = {
    "Student": ["Sidhi", "Shradhha", "Sneha", "Simran", "Chris"],
    "Marks": [88, 76, 91, 69, 95]
}

df = pd.DataFrame(student_data)

plt.bar(df["Student"], df["Marks"])

plt.title("Student Performance")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()

plt.hist(df["Marks"])

plt.title("Marks Distribution")

plt.show()
```
